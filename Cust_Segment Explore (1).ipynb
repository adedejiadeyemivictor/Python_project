```python
counts = Cust_seg_data[['Education Level']].value_counts().sort_index()

# Plot a bar chart
ax = counts.plot(kind='bar', figsize=(10, 4))

# Add data labels
for i in ax.patches:
    ax.text(i.get_x() + i.get_width() / 2, i.get_height() + 1, str(int(i.get_height())), ha='center', fontsize=9)

# Set plot labels and title
plt.title('Counts by Gender, Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')

# Show the plot
plt.show()
```
